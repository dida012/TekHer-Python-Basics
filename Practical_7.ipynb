### Exercise 4:
Plot a sine wave using Matplotlib for `x` values between 0 and 2.


```python
# Exercise 4
import matplotlib.pyplot as plt
import numpy as np

# Define the x-values
# We want x to range from 0 to 2*pi.
# np.linspace is perfect for creating a smooth curve.
# We'll use 1000 points to make the sine wave look very smooth.
x = np.linspace(0, 2 * np.pi, 1000)
# Define the y-values
# Calculate the sine of each x-value.
y = np.sin(x)
# Create the plot
plt.plot(x, y)
# Add labels and title
plt.xlabel("x (radians)")
plt.ylabel("sin(x)")
plt.title("Sine Wave (y = sin(x))")
#  Customize x-axis ticks to show multiples of pi (optional, but good for trig plots)
# Define the tick locations (e.g., 0, pi/2, pi, 3*pi/2, 2*pi)
tick_locations = [0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi]
# Define the labels for those ticks
tick_labels = ["0", "/2", "", "3/2", "2"]
plt.xticks(tick_locations, tick_labels)

# Add a grid and display the plot
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5) # Add x-axis line
plt.axvline(0, color='black', linewidth=0.5) # Add y-axis line
plt.show()
```

### Exercise 9:
Create a subplot with 2 rows and 1 column. Plot `y1 = x^2` in the first subplot and `y2 = x^3` in the second.


```python
# Exercise 9
import matplotlib.pyplot as plt
import numpy as np

#  Define the x-values
x = np.linspace(0, 10, 100)
#  Define the y-values for both functions
y1 = x**2  # y = x^2
y2 = x**3  # y = x^3
#  Create a subplot with 2 rows and 1 column
# `fig` is the overall figure, `axs` is an array of axes objects for each subplot.
# 2 rows, 1 column
fig, axs = plt.subplots(2, 1, figsize=(8, 8)) 
# Plot y1 = x^2 in the first subplot (axs[0])
axs[0].plot(x, y1, color='blue')
axs[0].set_title('Plot of y = x^2')
axs[0].set_xlabel('x-axis')
axs[0].set_ylabel('y-axis')
axs[0].grid(True)
# Plot y2 = x^3 in the second subplot (axs[1])
axs[1].plot(x, y2, color='red')
axs[1].set_title('Plot of y = x^3')
axs[1].set_xlabel('x-axis')
axs[1].set_ylabel('y-axis')
axs[1].grid(True)
# Add a title to the entire figure (optional)
fig.suptitle('Subplots of y = x^2 and y = x^3', fontsize=16)
#  Adjust layout to prevent overlapping titles/labels
# Adjust rect to make space for suptitle
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
#  Display the plot
plt.show()
```

### Exercise 16:
Create a stacked bar chart for the following data: `Group A = [3, 5, 7]`, `Group B = [2, 6, 4]`, and `categories = ['X', 'Y', 'Z']`.


```python
# Exercise 16
import matplotlib.pyplot as plt
import numpy as np

# Your data
categories = ['X', 'Y', 'Z']
 # Convert to numpy array for easy addition
group_a_values = np.array([3, 5, 7])
group_b_values = np.array([2, 6, 4])
# Create the figure and axes
plt.figure(figsize=(8, 6))
# Plot the bars for Group A (bottom layer)
plt.bar(categories, group_a_values, label='Group A', color='skyblue')
# Plot the bars for Group B (top layer), stacked on top of Group A
# This is the key for stacking!
plt.bar(categories, group_b_values, label='Group B', color='lightcoral',
        bottom=group_a_values) 
# Add labels, title, and legend
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Stacked Bar Chart of Group A and Group B")
# Show the legend to differentiate groups
plt.legend() 
# Add a grid for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
# Display the plot
plt.show()
```

### Exercise 17:
Create a plot with multiple subplots in a 2x2 grid. Plot `y1 = x`, `y2 = x^2`, `y3 = x^3`, and `y4 = x^4`.


```python
# Exercise 17
import matplotlib.pyplot as plt
import numpy as np

# Define the x-values
x = np.linspace(0, 10, 100)
# Define the y-values for each function
y1 = x
y2 = x**2
y3 = x**3
y4 = x**4
# Create a 2x2 subplot grid
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
# Plot each function in its respective subplot
axs[0, 0].plot(x, y1, color='blue')
axs[0, 0].set_title('y = x')

axs[0, 1].plot(x, y2, color='red')
axs[0, 1].set_title('y = x^2')

axs[1, 0].plot(x, y3, color='green')
axs[1, 0].set_title('y = x^3')

axs[1, 1].plot(x, y4, color='purple')
axs[1, 1].set_title('y = x^4')
# Add a title to the entire figure
fig.suptitle('Plots of y = x, x^2, x^3, and x^4', fontsize=16)
# Add labels and grids to all subplots
for ax in axs.flat:
    ax.set(xlabel='x-axis', ylabel='y-axis')
    ax.grid(True)
# Adjust layout to prevent overlapping
# Adjust rect to make space for suptitle
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust rect to make space for suptitle
# Display the plot
plt.show()
```

### Exercise 21:
Plot a contour plot for the function `z = sin(x) + cos(y)` over a grid of `x` and `y` values ranging from -5 to 5.


```python
# Exercise 21
import matplotlib.pyplot as plt
import numpy as np

#  Define the range for x and y
x_range = np.linspace(-5, 5, 100)
y_range = np.linspace(-5, 5, 100)
# Create a 2D grid of x and y values
X, Y = np.meshgrid(x_range, y_range)
#  Calculate the z-values using the function z = sin(x) + cos(y)
Z = np.sin(X) + np.cos(Y)
#  Create the contour plot
plt.figure(figsize=(8, 7))
# Use plt.contourf for filled contours
# levels controls the number of contour lines/regions
contour_plot = plt.contourf(X, Y, Z, levels=20, cmap='viridis') 
# Add a colorbar to interpret the z-values
plt.colorbar(contour_plot, label='z = sin(x) + cos(y)')
# Add labels and title
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Contour Plot of z = sin(x) + cos(y)")
# Display the plot
plt.show()
```

### Exercise 36:

Create a wind rose chart using random wind direction and speed data.


```python
# Exercise 42
import matplotlib.pyplot as plt
import numpy as np

#  Generate Random Wind Data 
# Number of wind observations
num_samples = 2000 
# Wind directions in degrees (0-360)
wind_directions_deg = np.random.rand(num_samples) * 360
# Wind speeds in arbitrary units (e.g., m/s or mph)
# Speeds from 0 to 20
wind_speeds = np.random.rand(num_samples) * 20 
# Convert directions to radians for Matplotlib's polar plot
wind_directions_rad = np.deg2rad(wind_directions_deg)
# Define Direction and Speed Bins
# Define the edges for direction bins (e.g., 8 cardinal directions)
# For N (0/360), NE (45), E (90), etc.
# We need edges for the histogram, so 9 edges for 8 bins (each 45 degrees wide)
direction_bin_edges_deg = np.arange(0, 360 + 45, 45)
# Define the center angles for plotting the bars (e.g., 0, 45, 90, ...)
direction_bin_centers_rad = np.deg2rad(np.arange(0, 360, 45))
# Define labels for the direction bins
direction_labels = ['N', 'NE', 'E', 'SE', 'S', 'SW', 'W', 'NW']
# Define speed bins (thresholds for speed categories)
speed_bins = [0, 5, 10, 15, 20] # Speeds: 0-5, 5-10, 10-15, 15-20
# Define labels for the speed categories (for the legend)
speed_labels = ['0-5 m/s', '5-10 m/s', '10-15 m/s', '15-20 m/s']
# Bin the Data and Calculate Frequencies
# Use np.histogram2d to count occurrences in each (direction bin, speed bin)
# The output 'counts' has shape (num_direction_bins, num_speed_bins-1)
counts, _, _ = np.histogram2d(
    wind_directions_deg,  # Data for the first dimension (x-axis of histogram2d)
    wind_speeds,          # Data for the second dimension (y-axis of histogram2d)
    bins=[direction_bin_edges_deg, speed_bins] # Bin edges for both dimensions
)
# Normalize the counts to percentages of the total number of observations
total_observations = num_samples
percentages = (counts / total_observations) * 100
#Create the Wind Rose Plot 
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Calculate the width of each bar in radians
# A bit less than the bin width (45 degrees = np.deg2rad(45)) to leave gaps
bar_width_rad = np.deg2rad(40)
# Initialize the 'bottom' of the stacked bars to zero for the first speed category
bottom_values = np.zeros(len(direction_bin_centers_rad))

# Define colors for the speed categories
# You can customize these colors
colors = ['#ADD8E6', '#87CEEB', '#4682B4', '#1E90FF'] # Light blue to dark blue
# Loop through each speed category to stack the bars
for i in range(len(speed_bins) - 1):
    # Plot bars for the current speed range
    ax.bar(
        direction_bin_centers_rad, # Angular positions of the bars (center of each direction bin)
        percentages[:, i],         # Heights of the bars (percentage for this speed bin)
        width=bar_width_rad,       # Width of each bar
        bottom=bottom_values,      # Base of the bars for stacking
        label=speed_labels[i],     # Label for the legend
        alpha=0.8,                 # Transparency
        color=colors[i],           # Custom color for this speed category
        edgecolor='white',         # White edge to separate bars
        linewidth=0.5
    )
    # Update the bottom_values for the next stack
    bottom_values += percentages[:, i]

# Customize the Plot 
# Set the angular tick labels (N, NE, E, etc.)
ax.set_xticks(direction_bin_centers_rad)
ax.set_xticklabels(direction_labels)
# Set radial (y-axis) ticks for percentage frequencies
# Determine the maximum percentage for radial axis limits
max_percentage = percentages.sum(axis=1).max()
# Calculate tick intervals (e.g., 5% if max is below 25%, 10% otherwise)
radial_tick_interval = 5 if max_percentage < 25 else 10
# Set the radial ticks and their labels
ax.set_yticks(np.arange(0, max_percentage + radial_tick_interval, radial_tick_interval))
ax.set_yticklabels([f'{p:.0f}%' for p in np.arange(0, max_percentage + radial_tick_interval, radial_tick_interval)])
ax.set_ylim(0, max_percentage + radial_tick_interval) # Ensure all ticks are visible
# Adjust the position of the radial tick labels (optional)
ax.set_rlabel_position(-22.5)
# Customize radial tick label appearance
ax.tick_params(axis='y', colors='dimgray', labelsize=8)
# Add title and legend
ax.set_title("Wind Rose Chart (Random Data)", va='bottom', fontsize=16)
ax.legend(title="Wind Speed", loc='lower left', bbox_to_anchor=(1.05, 0.0)) # Place legend outside
# Adjust layout to prevent clipping of legend
plt.tight_layout(rect=[0, 0, 0.95, 1]) # [left, bottom, right, top] normalized coordinates
plt.show()
```

### Exercise 43:
Plot a cumulative frequency curve for a dataset of 100 random numbers.


```python
# Exercise 43
import matplotlib.pyplot as plt
import numpy as np

# Generate a dataset with 100 random numbers (e.g., from a uniform distribution)
data = np.random.rand(100) * 100 # Random numbers between 0 and 100
# Sort the data
# Sorting is crucial for calculating cumulative frequency correctly.
data_sorted = np.sort(data)
# Calculate cumulative frequencies
# The cumulative frequency for each point is its index plus 1 (to start counting from 1)
# divided by the total number of points, giving a proportion.
cumulative_frequency = np.arange(1, len(data_sorted) + 1) / len(data_sorted)
# Create the cumulative frequency curve
plt.figure(figsize=(10, 6))
plt.plot(data_sorted, cumulative_frequency, marker='.', linestyle='-', color='purple', alpha=0.7)
#  Add labels, title, and grid
plt.xlabel("Data Value")
plt.ylabel("Cumulative Frequency (Proportion)")
plt.title("Cumulative Frequency Curve of 100 Random Numbers")
plt.grid(True, linestyle='--', alpha=0.6)
# Set y-axis limits to clearly show the 0 to 1 range of cumulative proportion
plt.ylim(0, 1.05)
# Extend x-limits slightly
plt.xlim(min(data_sorted) * 0.9, max(data_sorted) * 1.1) 
# Display the plot
plt.show()
```

### Exercise 44:
Create a waterfall chart using Matplotlib to show the changes in a sequence of values.


```python
# Exercise 44
import matplotlib.pyplot as plt
import numpy as np

#  Define your data 
# Initial value
initial_value = 100
# Changes (positive for increase, negative for decrease)
# e.g., Revenue: Start 100, +50 (sales), -20 (returns), +30 (new product), -15 (discount), Final
changes = [50, -20, 30, -15]
# Labels for each step (including start and end)
labels = ['Start', 'Sales', 'Returns', 'New Product', 'Discount', 'End']
#Calculate cumulative values and bar properties
# Calculate the running total for each step
running_total = [initial_value]
for change in changes:
    running_total.append(running_total[-1] + change)

final_value = running_total[-1]

# Prepare data for plotting
bar_heights = [initial_value] + changes + [final_value]
# Initialize all bottoms to 0
bar_bottoms = [0] * len(bar_heights) 
# Determine colors for each bar (green for positive, red for negative, blue for start/end)
colors = []
for i, height in enumerate(bar_heights):
    if i == 0 or i == len(bar_heights) - 1: # Start and End bars
        colors.append('steelblue')
    elif changes[i-1] > 0: # Positive change
        colors.append('forestgreen')
    else: # Negative change
        colors.append('firebrick')

# Calculate the `bottom` positions for floating bars
# The "bottom" of a change bar is the running total BEFORE that change occurred.
# The "bottom" of a final bar is 0.
bottoms = [0] # Start bar always starts at 0

for i in range(len(changes)):
    if changes[i] > 0:
        # For positive changes, bar starts at the previous running total
        bottoms.append(running_total[i])
    else:
        # For negative changes, bar starts at the (previous running total + change)
        # This makes the bar go downwards from the previous total
        bottoms.append(running_total[i+1])

bottoms.append(0) # Final bar also starts at 0

# Convert bar_heights to absolute values for plotting (heights must be positive)
bar_plot_heights = [abs(h) for h in bar_heights]

# Create the Waterfall Chart
fig, ax = plt.subplots(figsize=(12, 7))
# Plot the bars
bars_plotted = ax.bar(labels, bar_plot_heights, bottom=bottoms, color=colors,
                      width=0.7, edgecolor='black', linewidth=0.5)

# Add Connector Lines (Optional but common in waterfall charts) 
# We want lines between each bar's top and the next bar's bottom
for i in range(len(labels) - 1):
    x_start = i + 0.35 # Right edge of the current bar
    x_end = i + 0.65   # Left edge of the next bar

    # Y-coordinate is the top of the current bar (running_total)
    y_level = running_total[i]
    ax.plot([x_start, x_end], [y_level, y_level],
            color='gray', linestyle='--', linewidth=1)

# Add Annotations (Values on top of/below bars)
for i, bar in enumerate(bars_plotted):
    yval = bar.get_y() + bar.get_height() / 2 # Center of the bar for annotation
    if i == 0 or i == len(labels) - 1: # For Start and End bars, show the value
        ax.text(bar.get_x() + bar.get_width() / 2, yval,
                f'{bar_plot_heights[i]:.0f}', ha='center', va='center', color='white',
                fontsize=10, fontweight='bold')
    else: # For change bars, show the change value
        change_val = changes[i-1]
        text_color = 'white' if change_val > 0 else 'white' # Text color for positive/negative changes
        ax.text(bar.get_x() + bar.get_width() / 2, yval,
                f'{change_val:+.0f}', ha='center', va='center', color=text_color,
                fontsize=10, fontweight='bold')

# Customize the plot 
ax.set_ylabel("Value")
ax.set_title("Waterfall Chart: Changes in Value Over Time")
ax.grid(axis='y', linestyle='--', alpha=0.7)
ax.set_ylim(bottom=min(bottoms) - (max(bar_plot_heights)/10), top=max(running_total) * 1.1) # Adjust Y-limits
# Rotate x-axis labels if they overlap
plt.xticks(rotation=45, ha='right') 
 # Adjust layout to prevent labels from overlapping
plt.tight_layout()
plt.show()
```

### Exercise 46:
Create a line plot with different markers for each point.


```python
# Exercise 46
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data (a relatively small number of points for clarity)
num_points = 12
x = np.linspace(0, 2 * np.pi, num_points)
y = np.sin(x) + np.random.rand(num_points) * 0.2 # Add some noise

# Define a list of different marker styles
# Matplotlib has many marker styles: 'o', 'v', '^', '<', '>', 's', 'p', '*', 'h', 'H', '+', 'x', 'D', 'd', '|', '_'
markers = ['o', 's', '^', 'D', 'p', '*', 'v', 'X', 'P', 'h', '+', 'd'] # A list of 12 distinct markers

# Create the figure and plot the main line
plt.figure(figsize=(10, 6))
# Plot the line without markers, so the connection is clear
plt.plot(x, y, color='blue', linestyle='-', linewidth=1, alpha=0.7, label='Data Trend Line')

# Plot each point with a different marker
# We loop through the data points and apply a marker from our 'markers' list
# The modulo operator (%) ensures we cycle through the markers if num_points > len(markers)
for i in range(num_points):
    plt.plot(x[i], y[i], marker=markers[i % len(markers)],
             markersize=8, color='red', markeredgecolor='black',
             label=f'Point {i+1} ({markers[i % len(markers)]})' if i == 0 else "") # Only label the first point for legend clarity

#  Add labels, title, and legend
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot with Different Markers for Each Point")
plt.legend(title="Markers Used", loc='upper right') # Add a legend for clarity

#  Add a grid
plt.grid(True, linestyle=':', alpha=0.6)

#  Display the plot
plt.show()
```

### Exercise 47:
Create a time series plot using Matplotlib and plot random data over a range of dates.


```python
# Exercise 47
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd # Pandas is excellent for handling dates and time series

# Generate a range of dates 
start_date = '2023-01-01'
end_date = '2023-12-31'
# Create a daily date range for a full year
date_range = pd.date_range(start=start_date, end=end_date, freq='D')

# Generate random data for each date 
num_days = len(date_range)
np.random.seed(42) # Set a seed for reproducibility of random numbers
# Simulate data with a base value and cumulative random changes
random_changes = np.random.randn(num_days) * 0.5
random_data = 50 + np.cumsum(random_changes) # Data starting around 50

# Create the time series plot 
fig, ax = plt.subplots(figsize=(14, 7)) # Create a figure and an axes object for explicit control

# Plot the data using numerical indices (0 to num_days-1) on the x-axis.
# This avoids issues with Matplotlib's automatic date locators in some environments.
ax.plot(np.arange(num_days), random_data, color='teal', linestyle='-', linewidth=1.5, alpha=0.8)

#Customize the plot for time series with manual date labeling
ax.set_xlabel("Date")
ax.set_ylabel("Value (Random Data)")
ax.set_title("Time Series Plot of Random Data (Daily)")

# Manually determine positions for x-axis ticks
# We'll place ticks approximately every 3 months (90 days)
tick_interval = 90
tick_positions = np.arange(0, num_days, tick_interval)

# Ensure the last date is included as a tick if it's not already
if num_days - 1 not in tick_positions and num_days > 0:
    tick_positions = np.append(tick_positions, num_days - 1)
tick_positions = np.unique(tick_positions) # Remove any duplicate tick positions

# Format the corresponding date labels from the 'date_range' using the selected positions
tick_labels = [date_range[int(pos)].strftime('%Y-%m-%d') for pos in tick_positions]

# Apply the custom tick positions and labels to the x-axis
ax.set_xticks(tick_positions)
ax.set_xticklabels(tick_labels)

# Rotate the date labels for better readability, especially with longer date formats
plt.xticks(rotation=45, ha='right') # 'ha' means horizontal alignment

# Add a grid to the plot for easier data reading
ax.grid(True, linestyle=':', alpha=0.7)

# Adjust layout to prevent labels and title from overlapping
plt.tight_layout()

# Display the plot
plt.show()
```

### Exercise 49:
Create a radar chart to visualize the strengths of five different attributes.


```python
# Exercise 49
import matplotlib.pyplot as plt
import numpy as np

# Define the five attributes (e.g., skills, qualities)
attributes = ['Communication', 'Problem-Solving', 'Teamwork', 'Creativity', 'Leadership']
num_attributes = len(attributes)

#  Generate random 'strength' values for each attribute (e.g., on a scale of 1-10)
# Let's make sure the values are visible, so we'll start them at 3.
# np.random.rand(num_attributes) generates values between 0 and 1.
strength_values = np.random.rand(num_attributes) * 7 + 3 # Values will be between 3 and 10

# To close the radar chart shape, we need to append the first value to the end
strength_values = np.concatenate((strength_values, [strength_values[0]]))

#  Calculate the angle for each axis
# We need to distribute the attributes evenly around a circle (2*pi radians)
# The `endpoint=False` means 2*pi itself is not included, and then we manually add the first angle again
angles = np.linspace(0, 2 * np.pi, num_attributes, endpoint=False).tolist()
angles.append(angles[0]) # Complete the loop for plotting

# Create the radar chart figure
# `subplot_kw=dict(polar=True)` is essential for creating a polar plot
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Plot the strength values
# `ax.plot()` draws the line connecting the strength values
ax.plot(angles, strength_values, color='gold', linewidth=2, linestyle='solid', label='Attribute Strengths')
# `ax.fill()` shades the area enclosed by the line, making the chart more impactful
ax.fill(angles, strength_values, color='gold', alpha=0.35)

#  Customize the chart aesthetics

# Set the angular tick labels (the attribute names)
ax.set_xticks(angles[:-1]) # Use angles excluding the repeated last one
ax.set_xticklabels(attributes, fontsize=12, fontweight='bold')

# Set the radial (y-axis) limits and ticks for the strength scale
ax.set_ylim(0, 10) # Strengths are on a scale of 0 to 10
ax.set_yticks(np.arange(0, 11, 2)) # Display ticks at 0, 2, 4, 6, 8, 10
ax.tick_params(axis='y', colors='dimgray', labelsize=9) # Customize radial tick label appearance

# Position radial labels slightly offset from the axes for better readability
ax.set_rlabel_position(0) # Adjusts the angle at which the radial labels appear

# Add a title to the chart
ax.set_title('Attribute Strength Visualization', va='bottom', fontsize=16, pad=20)

# Add a legend
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1)) # Place legend slightly outside the plot

# Display the chart
plt.show()
```
